```python
from imblearn.over_sampling import SMOTE

X = df.drop(columns=['Class'])
y = df['Class']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print(pd.Series(y_resampled).value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

precision_fraud = precision_score(y_test, y_pred, pos_label=1)
precision_non_fraud = precision_score(y_test, y_pred, pos_label=0)

recall_fraud = recall_score(y_test, y_pred, pos_label=1)
recall_non_fraud = recall_score(y_test, y_pred, pos_label=0)

f1_fraud = f1_score(y_test, y_pred, pos_label=1)
f1_non_fraud = f1_score(y_test, y_pred, pos_label=0)

print("Accuracy:", accuracy)
print("Precision (Fraud):", precision_fraud)
print("Precision (Non-Fraud):", precision_non_fraud)
print("Recall (Fraud):", recall_fraud)
print("Recall (Non-Fraud):", recall_non_fraud)
print("F1-score (Fraud):", f1_fraud)
print("F1-score (Non-Fraud):", f1_non_fraud)
```
